```python
import numpy as np
import matplotlib.pyplot as plt

def discrete_filter(bel, d):
    bel_prime = np.zeros(bel.shape[0])

    for i in range(bel.shape[0]):
        if d == 1:  # Moving forward
            bel0 = bel[i]
            bel1 = bel[i - 1] if i - 1 >= 0 else 0
            bel2 = bel[i - 2] if i - 2 >= 0 else 0

            if i == bel.shape[0] - 1:  # Last cell
                bel_prime[i] = bel0 + bel1 + bel2
            elif i == bel.shape[0] - 2:  # Second to last cell
                bel_prime[i] = 0.75 * bel1 + 0.25 * bel2
                bel_prime[i + 1] = 0.25 * bel1
            else:
                bel_prime[i] += 0.25 * bel2 + 0.5 * bel1 + 0.25 * bel0

        elif d == -1:  # Moving backward
            bel0 = bel[i]
            bel1 = bel[i + 1] if i + 1 < bel.shape[0] else 0
            bel2 = bel[i + 2] if i + 2 < bel.shape[0] else 0

            if i == 0:  # First cell
                bel_prime[i] = bel0 + bel1 + bel2
            elif i == 1:  # Second cell
                bel_prime[i] = 0.75 * bel1 + 0.25 * bel2
                bel_prime[i - 1] = 0.25 * bel1
            else:
                bel_prime[i] += 0.25 * bel2 + 0.5 * bel1 + 0.25 * bel0

    # Normalize the belief to ensure it sums to 1
    bel_prime /= np.sum(bel_prime)
    return bel_prime

def plot_histogram(bel):
    plt.clf()  # Clear the figure instead of the axis
    plt.bar(range(bel.shape[0]), bel, width=0.5) #width=1.0
    plt.axis([0, bel.shape[0] - 1, 0, 1])
    plt.draw()
    plt.pause(1)

def main():
    # Initialize the belief: robot starts at position 10
    bel = np.hstack((np.zeros(10), 1, np.zeros(9)))

    plt.figure()
    plt.ion()
    plt.show()

    # Perform 9 consecutive move forward commands
    for i in range(9):
        plot_histogram(bel)
        bel = discrete_filter(bel, 1)
        print(f"Step {i + 1} (forward):")
        print(bel)

    # Perform 3 consecutive move backward commands
    for i in range(3):
        plot_histogram(bel)
        bel = discrete_filter(bel, -1)
        print(f"Step {i + 1} (backward):")
        print(bel)

    # Final plot
    plt.ioff()
    plt.show()

if __name__ == '__main__':
    main()
```
